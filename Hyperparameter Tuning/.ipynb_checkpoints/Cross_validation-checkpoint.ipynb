```python
import pandas as pd
from sklearn.model_selection import train_test_split,KFold,LeaveOneOut,cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
url="https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
df=pd.read_csv(url)
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
X=df.drop(columns=['medv'])
Y=df['medv']
```

```python
# Leave one out cross validation
```

```python
model=LinearRegression()
loo=LeaveOneOut()

score=cross_val_score(model,X,Y,cv=loo,scoring='neg_mean_squared_error')
mse_score=-score
print(mse_score.mean())
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
print("Hold out approach r2 score:",r2_score(Y_test,y_pred))
```

```python
# k-fold Cross Validation:
```

```python
kfold=KFold(n_splits=10,shuffle=True,random_state=10)
score2=cross_val_score(model,X_train,Y_train,cv=kfold,scoring="r2")

print(score2.mean())
```
